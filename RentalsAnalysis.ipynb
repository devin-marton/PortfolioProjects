## Pipeline

```python
# Building a full pipeline of the cleaning and transformations done above

# Getting the data
import pandas as pd
filename = '/content/rentals.csv'
rentals = pd.read_csv(filename)

# Drop the NaN values in 'price'
rentals.dropna(subset=['price'], inplace=True)

# Drop columns not needed
rentals = rentals.drop(['name', 'host_name','neighbourhood_group',
                        'last_review','license'],
                       axis = 1)

# Splitting data
# Creating a new attribute to be able to randomly select training data
#   -- This alows for a better sample of the data to be taken
import numpy as np

rentals['min_night_cat'] = pd.cut(rentals['minimum_nights'],
                               bins=[0.0,1.0,3.0,5.0,7.5,np.inf],
                               labels=[1,2,3,4,5])

# Splitting data using stratified sampling to get test and train data
from sklearn.model_selection import train_test_split

strat_train_set, strat_test_set = train_test_split(
    rentals, test_size=0.2,stratify=rentals['min_night_cat'], random_state=42
)

# Remove the min_night_cat attribute from both training and test sets

for set_ in (strat_train_set, strat_test_set):
  set_.drop('min_night_cat', axis=1, inplace=True)

# Drop the 'price' feature since it is the target and transformations and
#   further cleaningshould not be done on this.
rentals = strat_train_set.drop('price', axis = 1)

# rentals_labels now just the target 'price' data

rentals_labels = strat_train_set['price'].copy()
```

```python
# Full Preprocessing Pipeline

from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import FunctionTransformer
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_selector, make_column_transformer
from sklearn.base import BaseEstimator, TransformerMixin
from sklearn.metrics.pairwise import rbf_kernel
from sklearn.preprocessing import StandardScaler

# Create ClusterSimilarity Class

from sklearn.cluster import KMeans

class ClusterSimilarity(BaseEstimator, TransformerMixin):
  def __init__(self, n_clusters = 10, gamma = 1.0, random_state= None):
    self.n_clusters = n_clusters
    self.gamma = gamma
    self.random_state = random_state

  def fit(self, X, y= None, sample_weight= None):
    self.kmeans_ = KMeans(self.n_clusters, n_init=10,
                          random_state=self.random_state)
    self.kmeans_.fit(X, sample_weight=sample_weight)
    return self

  def transform(self, X):
    return rbf_kernel(X, self.kmeans_.cluster_centers_, gamma=self.gamma)

  def get_feature_names_out(self, names=None):
    return [f'Cluster {i} similarity' for i in range(self.n_clusters)]

################################################################################
# The categorical feature will be encoded
################################################################################

cat_pipeline=make_pipeline(
    SimpleImputer(strategy='most_frequent'),
    OneHotEncoder(handle_unknown='ignore')
)

################################################################################
# Features that should make logarithmic to create gausian distribution
################################################################################

log_pipeline=make_pipeline(
    SimpleImputer(strategy='median'),
    FunctionTransformer(np.log, feature_names_out='one-to-one'),
    StandardScaler()
)


################################################################################
# Latitude and Longitude features replaced by cluster similarity features
################################################################################

cluster_simil=ClusterSimilarity(n_clusters=10, gamma=.1, random_state=42)

################################################################################
# All the remaining numerical features are standardized
################################################################################

default_num_pipeline = make_pipeline(SimpleImputer(strategy='median'),
                                     StandardScaler())

################################################################################
# Put all the preprocessing together
################################################################################

preprocessing=ColumnTransformer([
    ('log', log_pipeline, ['reviews_per_month',
                           'calculated_host_listings_count',
                           'minimum_nights']),
    ('geo', cluster_simil, ['latitude', 'longitude']),
    ('cat', cat_pipeline, make_column_selector(dtype_include= object)),
], remainder=default_num_pipeline)

################################################################################
# Fitting the ColumnTransformer with the rentals data
################################################################################

rentals_prepared = preprocessing.fit_transform(rentals)
rentals_prepared.shape

```

# Fine-Tune Model (Hyperparameters)

```python
from sklearn.experimental import enable_halving_search_cv
from sklearn.model_selection import HalvingRandomSearchCV
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import randint

from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestRegressor

full_pipeline = Pipeline([
    ('preprocessing', preprocessing),
    ('random_forest', RandomForestRegressor(n_estimators=10,random_state=42))
])
# Hyerparameters used were the n_clusters in the geo function and max_features
#   in the random_forest function.
# These are being optimized with Random Search coming up with values:
#   - n_clusters: 4 and 23 both had same RMSE
#   - max_features: 2 and 8 both had the same RMSE
param_distribs = {'preprocessing__geo__n_clusters':randint(low=3, high=50),
                  'random_forest__max_features':randint(low=2, high=20)}

rnd_search = RandomizedSearchCV(
    full_pipeline, param_distributions=param_distribs, n_iter=10, cv=3,
    scoring='neg_root_mean_squared_error', random_state=42
)

rnd_search.fit(rentals, rentals_labels)
```

```python
# Displays the random search results

cv_res = pd.DataFrame(rnd_search.cv_results_)
cv_res.sort_values(by='mean_test_score', ascending=False, inplace=True)

cv_res=cv_res[['param_preprocessing__geo__n_clusters',
               'param_random_forest__max_features', 'split0_test_score',
               'split1_test_score', 'split2_test_score', 'mean_test_score']]

score_cols = ['split0', 'split1', 'split2', 'mean_test_rmse']
cv_res.columns=['N-clusters', 'max_features'] + score_cols
cv_res[score_cols] = -cv_res[score_cols].round().astype(np.int64)
cv_res.head(10)
```

```python
# Find the most important Features

final_model = rnd_search.best_estimator_
feature_importances = final_model['random_forest'].feature_importances_
sorted(zip(feature_importances,
           final_model['preprocessing'].get_feature_names_out()),
       reverse=True)
```

```python
from geopy.geocoders import Nominatim
import time

geolocator = Nominatim(user_agent='locations')

def get_address(lat, lon):
    try:
        location = geolocator.reverse((lat, lon), exactly_one=True)
        if location:
            return location.address
        return "No address found"
    except Exception as e:
        return f"Error: {str(e)}"

# Create a new column for addresses
rentals['address'] = None

max_rows = min(10, len(rentals))
processed_rentals = rentals.head(max_rows).copy()

for index, row in processed_rentals.iterrows():
    address = get_address(row['latitude'], row['longitude'])
    processed_rentals.at[index, 'address'] = address
    print(f"Processed {index+1}/{max_rows}: {address}")

    time.sleep(1)

# Display the results
print("\nFirst 10 Addresses:")
print(processed_rentals[['latitude', 'longitude', 'address']])
```
